```python
import pandas as pd
df=pd.read_csv('Tweets.csv')
df.iloc[5,10]
```

```python
import nltk
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('state_union')
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')
nltk.download('movie_reviews')
nltk.download('punkt')
nltk.download('maxent_treebank_pos_tagger')
```

```python
from nltk import sent_tokenize,word_tokenize
from nltk.corpus import stopwords
import string
from nltk.stem import WordNetLemmatizer
```

```python
words=[(df.iloc[i,10],df.iloc[i,1]) for i in range(14640)]
words[0:5]
```

```python
def tokenize(word):
    token_words=word_tokenize(word)
    return token_words

new_words=[(tokenize(word),category) for word,category in words]
print(new_words[0:5])
```

```python
stop_words=stopwords.words('english')
punct=list(string.punctuation)
stop_words+=punct
stop_words
```

```python
from nltk.stem import WordNetLemmatizer
from nltk import pos_tag
from nltk.corpus import state_union
lemmatizer=WordNetLemmatizer()

from nltk.corpus import wordnet
def get_simple_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN

def clean_review(words):
    output=[]
    for w in words:
        if(w.lower() not in stop_words):
            pos=pos_tag([w])
            clean_word=lemmatizer.lemmatize(w,pos=get_simple_pos(pos[0][1]))
            output.append(clean_word)
    return output
```

```python
documents=[(clean_review(word),category) for word,category in new_words]
```

```python
documents[5]
```

```python
training_data=documents[0:10000]
testing_data=documents[10000:]
```

```python
all_words=[]
for doc in training_data:
    all_words+=doc[0]
    
freq=nltk.FreqDist(all_words)
common=freq.most_common(10000)
features=[i[0] for i in common]
features
```

```python
def get_dict(words):
    new_dict={}
    for w in features:
        new_dict[w]=w in words
    return new_dict
```

```python
training_check=[(get_dict(words),category) for words,category in training_data]
testing_check=[(get_dict(words),category) for words,category in testing_data]
```

```python
testing_check[5]
```

```python
from nltk import NaiveBayesClassifier
classifier=NaiveBayesClassifier.train(training_check)
```

```python
nltk.classify.accuracy(classifier,testing_check)
```
